```python

import numpy as np 
import pandas as pd 

file_path="/kaggle/input/datasets/oleksiimartusiuk/e-commerce-data-shein/us-shein-appliances-3987.csv" 
df = pd.read_csv(file_path)

df.fillna("Unknown", inplace=True)

df['price'] = pd.to_numeric(df['price'].astype(str).str.replace('$', '', regex=False), errors='coerce')
df['discount'] = pd.to_numeric(df['discount'].astype(str).str.replace('%', '', regex=False).str.replace('-', '', regex=False), errors='coerce')

df['price'] = df['price'].fillna(df['price'].median())
df['discount'] = df['discount'].fillna(0)

df.drop_duplicates(inplace=True)
df['goods-title-link--jump'] = df['goods-title-link--jump'].astype(str).str.strip()

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['price'] >= (Q1 - 1.5 * IQR)) & (df['price'] <= (Q3 + 1.5 * IQR))]

df.to_csv('cleaned_shein_data.csv', index=False) #Save the clean file
```
